```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt
import seaborn as sns

import folium
from folium.plugins import HeatMap

from IPython.display import display
pd.set_option('display.max_colwidth', None)
```

```python
df = pd.read_csv('asset/weather_cleaned.csv')
df.head()
```

```python
#variance between station 1 and 2
```

```python
df['date'] = pd.to_datetime(df.Date)
df['year'] = df.date.dt.year
df['month'] = df.date.dt.month
df['weekday'] = df.date.dt.weekday
```

```python
df_temp1 = df.loc[df.Station == 1, :]
df_temp2 = df.loc[df.Station == 2, :]

feature = df.columns[4:]
```

Water temperature is one of the most important factors that account for development of larvae$^{1}$. Therefore we will pay more importance on the temperature feature

```python
fig, ax = plt.subplots(nrows=len(feature), ncols=1, figsize=(30,140))
for row in range(15):
    sns.lineplot(x='date', y=feature[row], data=df_temp1, label='station 1', ax=ax[row])
    sns.lineplot(x='date', y=feature[row], data=df_temp2, label='station 2', ax=ax[row])
    ax[row].tick_params(axis='both', which='major', labelsize=20)
    ax[row].set_ylabel(feature[row], fontsize=20)
    ax[row].set_xlabel("")
```

We can see that overall there is not much different in terms of station 1 and station 2 especially on the temperature feature. We can drop station 2 and proceed analysis with station 1

```python
df_train = pd.read_csv('asset/train_cleaned.csv')
df_train
```

```python
df_train['date'] = pd.to_datetime(df_train.date)
```

```python
df_train['year'] = df_train.date.dt.year
df_train['month'] = df_train.date.dt.month
df_train['weekday'] = df_train.date.dt.weekday
```

```python
series = df.groupby('month')['Tavg'].mean()
```

```python
ax = df_train.groupby('month').wnvpresent.sum().plot.bar(color='blue')
ax1 = ax.twinx()
df.groupby('month')['Tavg'].mean().plot(ax=ax1, use_index=False, color='red');
ax.set_title('Month / Temperature vs Virus Count')
ax.set_ylabel('Virus Count')
ax1.set_ylabel('Average Temperature');
```

From graph above, there is an indication of correlation between temperature & virus count. The 'phase' lag between temperature & virus count suggest the possibility of mosquitor breeding life cycle.

**Feature Engineering:**
   1. Create a month feature
   2. Create a temperature feature based on $T = T_{-15days}$ based on mosquitor life cycle

```python
ax = df_train.groupby(['species']).wnvpresent.value_counts(normalize=True).unstack().plot.barh(color = ['blue', 'red'])
ax.set_title('Specifies vs Virus Count')
ax.legend(bbox_to_anchor=(1.0, 1.05), title='virus presence');
```

Culex Pipiens & Culex Restuans are the species that have virus with the formal having higher %.

**Feature Engineering:**
   1. Create 3 groups with monotonic relationship with virus count. ie group 0: Culex pipiens, group 1: culex restuans, group 2: others

```python
def plotheatmap(time):
    map_virus = folium.Map(location=(df_train.latitude[0],df_train.longitude[0]), zoom_start=13,tiles="Stamen Terrain")

    virus_latlog = [[x, y] for x,y, virus, time 
                    in zip(df_train.latitude, df_train.longitude, df_train.wnvpresent, df_train[time])
                    if virus == 1]
    HeatMap(virus_latlog).add_to(map_virus)
    
    return map_virus
```

```python
plotheatmap('year')
```

A map of where the present of virus. We have explored the location of none virus obsercation. However it is found that the presence of virus does not have a correlation with geography
